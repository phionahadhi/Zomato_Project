```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

## Joining The Country-Code.xlsx file to the Zomato.csv file 

```python
# we can use the country code to join the teo tables and get the country name
final_df = pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

```python
# checking data types of the columns
final_df.dtypes
```

```python
final_df[['Currency']]
```

### Country Analysis

```python
country_names = final_df.Country.value_counts().index
```

```python
country_vals = final_df.Country.value_counts().values
```

## Pie Chart

```python
# pie chart - top 3 countries that uses zomatoe
plt.pie(country_vals[:3], labels = country_names[:3], autopct='%1.2f%%')
```

### Observations

Zomato maximum records and transactions are from India and 

second largest records are from United states then followed by United kingdom

## Numerical Variables

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
# Observations
# 1. When the rating is between 4.5 to 4.9, the rating color is dark green and the rating text is Excellent.
# 2. When the rating is between 4.0 to 4.4, the rating color is green and the rating text is Very Good.
# 3. When the rating is between 3.5 to 3.9, the rating color is light green and the rating text is Good.
# 4. When the rating is between 3.0 to 3.4, the rating color is yellow and the rating text is Average.
# 5. When the rating is between 2.5 to 2.9, the rating color is orange and the rating text is Poor.
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count', hue = 'Rating color', data=ratings, palette=['blue', 'red', 'orange','yellow','lightgreen','darkgreen'])
```

Observations

1. Not rated count is very high

2. Maximum number of rating are between 2.5 to 3.4

```python
# count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue', 'red', 'orange','yellow','lightgreen','darkgreen'])
```

```python
# Find the countries that has given 0 rating
zero_rating = final_df[final_df['Aggregate rating']==0].groupby('Country').size().reset_index().rename(columns={0:'Zero Rating Count'}).sort_values('Zero Rating Count',ascending=False)
zero_rating
```

Observation: Maximum number of 0 ratings are from Indian customers
